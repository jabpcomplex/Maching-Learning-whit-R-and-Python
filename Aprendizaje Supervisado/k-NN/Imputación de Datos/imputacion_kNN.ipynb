```python
df = pd.read_csv("base_SIN_IMPUTAR.csv", encoding='iso8859-1')
```

```python
df.head()
```

```python
datos = df.iloc[:,[0]+list(range(9, 230))] ; datos.head(10)
```

```python
from sklearn.impute import KNNImputer
```

```python
imputer = KNNImputer(n_neighbors=6, weights="distance")
imputados_uni = imputer.fit_transform(datos)
datos_imputados = pd.DataFrame(imputados_uni)
#np.stack((imputados_uni, imputados_dis), axis=1)
```
